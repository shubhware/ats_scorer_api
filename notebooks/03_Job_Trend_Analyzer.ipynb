```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import os

print("1. Loading the massive LinkedIn Job Postings dataset...")
file_path = "../data/job_postings/postings.csv"

df = pd.read_csv(file_path, usecols=['title', 'description']).dropna()

print("\n2. Filtering out Sales/Retail to focus on Engineering & Tech...")
tech_keywords = 'software|developer|engineer|data|cyber|security|cloud|it|network|systems|ai|machine learning|web'
tech_df = df[df['title'].str.contains(tech_keywords, case=False, na=False)].head(50000)

# --- THE MASSIVE TECH DICTIONARY ---
target_skills = [
    # Core Languages
    "python", "java", "c++", "c", "c#", "javascript", "typescript", "go", "golang", "rust", "ruby", "php", "swift", "kotlin", "r", "dart",
    
    # Web Development (Frontend & Backend)
    "react", "angular", "vue", "next.js", "node.js", "express", "django", "flask", "spring boot", "html", "css", "tailwind", "bootstrap", "sass", "graphql", "rest api", "mern", "mean", "laravel", "asp.net", "web development",
    
    # Cyber Security & Networking
    "cyber security", "information security", "penetration testing", "ethical hacking", "nmap", "metasploit", "wireshark", "kali linux", "cryptography", "owasp", "burp suite", "firewall", "malware analysis", "incident response", "soc", "splunk", "siem", "vulnerability assessment", "cissp", "ceh", "identity and access management", "iam", "network security",
    
    # Cloud, DevOps & Infrastructure
    "aws", "azure", "gcp", "google cloud", "docker", "kubernetes", "jenkins", "ci cd", "ci/cd", "terraform", "ansible", "linux", "ubuntu", "bash", "shell scripting", "git", "github", "gitlab", "bitbucket", "nginx", "apache", "microservices",
    
    # Data Science, AI & Machine Learning
    "machine learning", "deep learning", "artificial intelligence", "ai", "nlp", "natural language processing", "computer vision", "tensorflow", "pytorch", "scikit-learn", "pandas", "numpy", "matplotlib", "data analysis", "data engineering", "big data", "hadoop", "spark", "databricks", "llm", "generative ai",
    
    # Databases & Storage
    "sql", "mysql", "postgresql", "mongodb", "redis", "oracle", "sql server", "cassandra", "elasticsearch", "dynamodb", "firebase", "nosql",
    
    # Core CS, DSA & Architecture
    "data structures", "algorithms", "dsa", "system design", "object oriented programming", "oop", "distributed systems", "agile", "scrum", "software engineering",
    
    # UI/UX & Testing
    "ui ux", "ui/ux", "figma", "adobe xd", "wireframing", "prototyping", "selenium", "cypress", "jest", "junit", "qa testing"
]

print("\n3. Initializing AI Trend Analyzer (Targeted Vocabulary)...")
vectorizer = TfidfVectorizer(
    vocabulary=target_skills,  
    ngram_range=(1, 2)         
)

print("4. Crunching the job market data...")
tfidf_matrix = vectorizer.fit_transform(tech_df['description'].str.lower())

print("\n5. Extracting the Top Trending Skills...")
feature_names = vectorizer.get_feature_names_out()
sums = tfidf_matrix.sum(axis=0)

data = []
for col, term in enumerate(feature_names):
    if sums[0, col] > 0:
        data.append((term, sums[0, col]))

ranking_df = pd.DataFrame(data, columns=['Technical Skill', 'Market Demand Score'])
ranking_df = ranking_df.sort_values('Market Demand Score', ascending=False)

print("\n--- 📈 TOP 25 TRENDING TECHNICAL SKILLS ---")
print(ranking_df.head(25).to_string(index=False))

save_dir = "../data/trends"
os.makedirs(save_dir, exist_ok=True)
ranking_df.to_csv(f"{save_dir}/top_market_skills.csv", index=False)
print(f"\nSuccess! Expanded market trends saved to {save_dir}/top_market_skills.csv")
```
