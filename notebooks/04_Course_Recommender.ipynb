```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer
import os

print("1. Loading Coursera Dataset...")
# Adjust the filename if yours is slightly different
df = pd.read_csv("../data/courses/coursera_courses.csv")

# Clean the data: Keep only the columns we actually need for the web app
# Note: If your CSV uses different column names (like just 'name' or 'description'), update them here!
df = df[['course_title', 'course_organization', 'course_skills', 'course_rating', 'course_url']].dropna()

print(f"Loaded {len(df)} professional courses.")

# Combine the title and skills into one rich text string for the AI to read
df['combined_text'] = df['course_title'] + " " + df['course_skills']

print("\n2. Booting up the Master Brain (all-MiniLM-L6-v2)...")
model = SentenceTransformer('all-MiniLM-L6-v2')

print("3. Converting thousands of courses into mathematical vectors...")
print("This will take about 15-30 seconds on your M3 chip.")
# Convert the combined text into embeddings
course_embeddings = model.encode(df['combined_text'].tolist(), convert_to_tensor=True, show_progress_bar=True)

print("\n4. Saving the AI Course Database...")
save_dir = "../app/core/models/course_vectors"
os.makedirs(save_dir, exist_ok=True)

# Save the raw data so the web app can read the titles and URLs
df.to_pickle(f"{save_dir}/course_catalog.pkl")

# Save the massive tensor of mathematical vectors
torch.save(course_embeddings, f"{save_dir}/course_embeddings.pt")

print(f"Success! Model 04 is built. Course Database saved to {save_dir}")
```
